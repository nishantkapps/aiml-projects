```python
dim2 = img2.shape
height2 = img2.shape[0]
width2 = img2.shape[1]
channels2 = img2.shape[2]

print('Image Dimension :', dim2)
print('Height :', height2)
print('Width :', width2)
print('Channels :', channels2)
```

```python
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1_rgb)
```

```python
plt.imshow(img2_rgb)
```

```python
# Extract Red Channel
img1_r = img1_rgb.copy()
img1_r[:,:,(1,2)]=0
plt.imshow(img1_r)
plt.title('Red Channel Image')
```

```python
# Resize is width, height
# Image size is always height, width
img1_resize = cv2.resize(img1_rgb,(500,750))
plt.imshow(img1_resize)
```

```python
img2_resize = cv2.resize(img2_rgb,(500,750))
plt.imshow(img2_resize)
```

```python
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img1_grayr = cv2.resize(img1_gray,(500,750))
img2_grayr = cv2.resize(img2_gray,(500,750))
```

```python
sum_image = cv2.addWeighted(img1_grayr,0.9,img2_grayr,0.5,0)
plt.imshow(sum_image, cmap='gray')
```

```python
sub_image=cv2.subtract(sum_image,100) #
plt.imshow(sub_image,cmap='gray')
plt.title('Subtract')
plt.show()
```

```python
# Image blurring or smoothing (introducing noise into the image)
# - apply filter to the image, convolutional operation
```

```python
plt.imshow(img1_rgb)
plt.title('RGB Image')
plt.show()

gaussian = cv2.GaussianBlur(img1_rgb,(7,7),0)
plt.imshow(gaussian)
plt.title('Gaussian Blur')
plt.show()

median = cv2.medianBlur(img1_rgb,35)
plt.imshow(median)
plt.title('Median Filter')
plt.show()
```

```python
# Edge detection in Images
# - determine sharp edges in source images
# Canny edge detection algorithm used to find edges in an image
```

```python
# Review Canny Edge Detection algorithm to understand more about it
# http://en.wikipedia.org/wiki/Canny_edge_detector
image_canny=cv2.Canny(img1_rgb, 100, 100, apertureSize=3)
plt.imshow(image_canny)
plt.title('Image Detection')
```
