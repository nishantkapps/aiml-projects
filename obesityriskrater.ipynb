```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/playground-series-s4e2/train.csv')
```

```python
df.head()
```

```python
cols = df.columns
num_cols = list(df._get_numeric_data().columns)
cat_cols = list(set(cols) - set(num_cols))
```

```python
df['FAVC'] = df['FAVC'].replace({'yes': 1, 'no': 0})
df['SMOKE'] = df['SMOKE'].replace({'yes': 1, 'no': 0})
df['SCC'] = df['SCC'].replace({'yes': 1, 'no': 0})
df['family_history_with_overweight'] = df['family_history_with_overweight'].replace({'yes': 1, 'no': 0})
df['MTRANS'] = df['MTRANS'].replace({'Automobile': 0,'Motorbike':1,'Public_Transportation': 2,'Walking': 3,'Bike':4})
df['Gender'] = df['Gender'].replace({'Female':1, 'Male':0})
df['NObeyesdad'] = df['NObeyesdad'].replace({'Insufficient_Weight':0, 'Normal_Weight':1, 'Overweight_Level_I':2, 'Overweight_Level_II':3, 'Obesity_Type_I':4, 'Obesity_Type_II':5, 'Obesity_Type_III':6 })
df['CALC'] = df['CALC'].replace({'no':0, 'Sometimes':1, 'Frequently':2})
df['CAEC'] = df['CAEC'].replace({'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3})
```

```python
x = df.iloc[:,1:-1]
y = df.iloc[:,-1]
```

```python
xnorm=x.apply(lambda x: (x-x.mean())/ x.std(), axis=0)
```

```python
dftest = pd.read_csv('/kaggle/input/playground-series-s4e2/test.csv')
```

```python
dftest['FAVC'] = dftest['FAVC'].replace({'yes': 1, 'no': 0})
dftest['SMOKE'] = dftest['SMOKE'].replace({'yes': 1, 'no': 0})
dftest['SCC'] = dftest['SCC'].replace({'yes': 1, 'no': 0})
dftest['family_history_with_overweight'] = dftest['family_history_with_overweight'].replace({'yes': 1, 'no': 0})
dftest['MTRANS'] = dftest['MTRANS'].replace({'Automobile': 0,'Motorbike':1,'Public_Transportation': 2,'Walking': 3,'Bike':4})
dftest['Gender'] = dftest['Gender'].replace({'Female':1, 'Male':0})
dftest['CALC'] = dftest['CALC'].replace({'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3})
dftest['CAEC'] = dftest['CAEC'].replace({'no':0, 'Sometimes':1, 'Frequently':2, 'Always':3})
```

```python
dftestnorm=dftest.iloc[:,1:].apply(lambda x: (x-x.mean())/ x.std(), axis=0)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(xnorm, y, train_size=0.8, random_state=42)
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
import xgboost as xgb
```

```python
cls = xgb.XGBClassifier()
```

```python
cls.fit(xtrain, ytrain)
```

```python
ypred = cls.predict(xtest)
```

```python
cls.feature_importances_
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(ypred, ytest)
```

```python
predictions = cls.predict(dftestnorm)
```

```python
submission = pd.DataFrame({'id': dftest.id, 'NObeyesdad': predictions})
submission['NObeyesdad'] = submission['NObeyesdad'].replace({0:'Insufficient_Weight', 1:'Normal_Weight', 2:'Overweight_Level_I', 3:'Overweight_Level_II', 4:'Obesity_Type_I', 5:'Obesity_Type_II', 6:'Obesity_Type_III'})
submission.to_csv('submission.csv', index=False)
```
